In this project I will be working on a fake advertising dataset, indicating whether or not a particular internet user clicked on an Advertisement. We will try to create a model that will predict whether or not they will click on an add based off the features of that user.

The dataset contains the following features:
a). 'Daily Time Spent on Site': consumer time on site in minutes
b). 'Age': customer age in years
c). 'Area Income': average income of geographical area of customers
d). 'Daily Internet Usage': average minutes a day the consumer is on the internet
e). 'Ad Topic Line': headline of the advertisement
f). 'City': city of consumer
g). 'Male': whether or not consumer was male
h). 'Country': country of consumer
i). 'Timestamp': time on which consumer clicked on ad or closed window
j). 'Clicked on Ad': 0 or 1 indicating clicking on ad

```python
from sklearn.cross_validation import train_test_split
```

```python
X= ad_data[['Daily Time Spent on Site','Age','Area Income','Daily Internet Usage','Male']]
y= ad_data['Clicked on Ad']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train, y_train)
```

```python
predictions= logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```
